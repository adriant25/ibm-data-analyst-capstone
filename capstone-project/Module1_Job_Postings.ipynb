# Module 1 — Job Postings Analysis
**Source:** `job-postings.xlsx` — collected via the Stack Overflow Jobs API  
**Purpose:** Identify which cities have the highest demand for technology jobs.  
**Referenced by:** `Capstone_Survey_Analysis.ipynb` — Step 9, Slide 7

```python
import pandas as pd
import matplotlib.pyplot as plt

DATA_FILE = 'job-postings.xlsx'

# -- Load data ----------------------------------------------------------------
df = pd.read_excel(DATA_FILE)
print(f'Rows: {len(df)}')
df
```

```python
# -- Bar chart: job postings by location, descending -------------------------
df_sorted = df.sort_values('Number of Jobs', ascending=False)

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(df_sorted['Location'], df_sorted['Number of Jobs'],
              color='#4C72B0', edgecolor='white')
ax.set_title('Job Postings by Location', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Location', fontsize=11)
ax.set_ylabel('Number of Job Postings', fontsize=11)
ax.bar_label(bars, padding=4, fontsize=9)
ax.set_xticklabels(df_sorted['Location'], rotation=20, ha='right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('charts/job_postings.png', dpi=150, bbox_inches='tight')
plt.show()
print('Chart saved: charts/job_postings.png')
```

### Findings
- All 7 surveyed locations report the same number of job postings (**27,005**), suggesting the API returned uniform data across cities for this query.
- Los Angeles, New York, San Francisco, Washington DC, Seattle, Austin, and Detroit are the target markets analyzed.
- The uniform count indicates broad, nationwide demand for technology roles — no single city shows a clear concentration advantage in this dataset.
